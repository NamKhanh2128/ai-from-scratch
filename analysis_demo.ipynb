# Phân tích dữ liệu mẫu
Notebook này minh họa quy trình:
1. Tạo một file CSV chứa dữ liệu bán hàng.
2. Đọc file CSV và hiển thị dưới dạng bảng.
3. Vẽ biểu đồ cột so sánh doanh thu.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Tạo DataFrame mẫu
data = {
    "Tháng": ["T1", "T2", "T3", "T4", "T5", "T6"],
    "Doanh thu (triệu đồng)": [120, 135, 148, 162, 155, 170],
    "Chi phí (triệu đồng)": [80, 85, 90, 95, 88, 92]
}
df = pd.DataFrame(data)

# Lưu thành file CSV
df.to_csv("sales_data.csv", index=False)
print("Đã tạo file sales_data.csv")
```

```python
# Đọc file CSV vừa tạo
df_loaded = pd.read_csv("sales_data.csv")

# Hiển thị bảng dữ liệu
df_loaded
```

```python
# Vẽ biểu đồ cột so sánh doanh thu và chi phí
plt.figure(figsize=(8, 5))
x = range(len(df_loaded["Tháng"]))
width = 0.35

plt.bar([i - width/2 for i in x], df_loaded["Doanh thu (triệu đồng)"], width, label="Doanh thu", color="skyblue")
plt.bar([i + width/2 for i in x], df_loaded["Chi phí (triệu đồng)"], width, label="Chi phí", color="salmon")

plt.xlabel("Tháng")
plt.ylabel("Triệu đồng")
plt.title("Doanh thu và Chi phí theo tháng")
plt.xticks(x, df_loaded["Tháng"])
plt.legend()
plt.show()
```
