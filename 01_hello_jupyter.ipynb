```python
import sys
import numpy as np
import matplotlib.pyplot as plt

print("Python:", sys.version)
print("NumPy:", np.__version__)

# Vẽ đồ thị
x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)
plt.plot(x, y)
plt.title("Sine Wave")
plt.show()
```
